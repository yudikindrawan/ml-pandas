```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

years = np.arange(2010, 2015)
population = np.around(np.random.uniform(1, 5, 30), 1)
```

```python
population.shape, population[:5]
```

```python
years.shape, years, type(years), years.dtype
```

```python
# Line Plot
fig, ax = plt.subplots()
ax.plot(years, population[:5])
ax.set_xticks(years)
ax.set_xticklabels(years.astype(int))
ax.set(title="Pertumbuhan Populasi")
plt.show()
```

```python
# Bar Plot 
fig, ax =  plt.subplots()
ax.bar(years, population[:5])
ax.set(title="Pertumbuhan Populasi")
plt.show()
```

```python
# Scatter Plot
fig, ax = plt.subplots()
ax.scatter(years, population[:5])
ax.set(title="Pertumbuhan Populasi")
plt.show()
```
