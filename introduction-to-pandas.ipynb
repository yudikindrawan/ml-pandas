```python
# Create a column from a pandas Series
seats_column = pd.Series([5,5,5,5,5])
car_sales["Seats"] = seats_column
car_sales
```

```python
# Create column from a pandas series
car_sales["Seats"] = car_sales["Seats"].fillna(5)
car_sales
```

```python
# Create column from a python list, note : if use python list lenght list must be same with lenght index of data
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"] * car_sales["Fuel per 100KM"] / 100
car_sales
```

```python
# Create a colum from a single value
car_sales["Number of wheels"] = 4
car_sales
```
