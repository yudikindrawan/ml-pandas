```python
import numpy as np
import matplotlib.pyplot as plt

# Buat Dataset Nilai Ujian (Mean=75, Std Dev=10)
np.random.seed(42)
nilai_ujian = np.random.normal(loc=75, scale=10, size=100)

# Hitung Mean, Variance, dan Standard Deviation
mean_nilai = np.mean(nilai_ujian)
var_nilai = np.var(nilai_ujian)
std_nilai = np.std(nilai_ujian)

# Tampilkan Hasil Perhitungan
print(f"Mean: {mean_nilai:.2f}")
print(f"Varians: {var_nilai:.2f}")
print(f"Standar Deviasi: {std_nilai:.2f}")

# Tentukan batas sigma
# bound_1sigma = (mean_nilai - std_nilai, mean_nilai + std_nilai)
# bound_2sigma = (mean_nilai - 2*std_nilai, mean_nilai + 2*std_nilai)
# bound_3sigma = (mean_nilai - 3*std_nilai, mean_nilai + 3*std_nilai)
bound_1sigma = (mean_nilai - std_nilai, mean_nilai + std_nilai)
bound_2sigma = (mean_nilai - 2*std_nilai, mean_nilai + 2*std_nilai)
bound_3sigma = (mean_nilai - 3*std_nilai, mean_nilai + 3*std_nilai)
# Tampilkan sigma
print(f"Sigma ±1 Std Dev: {bound_1sigma[0]:.2f}%, {bound_1sigma[1]:.2f}%")
print(f"Sigma ±2 Std Dev: {bound_2sigma[0]:.2f}%, {bound_2sigma[1]:.2f}%")
print(f"Sigma ±3 Std Dev: {bound_3sigma[0]:.2f}%, {bound_3sigma[1]:.2f}%")

# Visualisasi Histogram
plt.figure(figsize=(10,5))
plt.hist(nilai_ujian, bins=20, alpha=0.7, color='blue', edgecolor='black', density=True)

# Tambahkan Garis Mean dan Standar Deviasi
plt.axvline(mean_nilai, color='red', linestyle='dashed', linewidth=2, label="Mean")
plt.axvline(mean_nilai - std_nilai, color='green', linestyle='dotted', linewidth=2, label="-1 Std Dev")
plt.axvline(mean_nilai + std_nilai, color='green', linestyle='dotted', linewidth=2, label="+1 Std Dev")

plt.legend()
plt.title("Distribusi Nilai Ujian Siswa")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.show()
```

```python
within_1sigma = np.sum((nilai_ujian >= mean_nilai - std_nilai) & 
                       (nilai_ujian <= mean_nilai + std_nilai)) / len(nilai_ujian) * 100
print(f"Persentase dalam ±1 Std Dev: {within_1sigma:.2f}%")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
hr_scores = np.random.normal(loc=75, scale=3, size=50)
it_scores = np.random.normal(loc=72, scale=8, size=50)
sales_scores = np.random.normal(loc=65, scale=15, size=50)

df = pd.DataFrame({
    'Departement' : ['HR'] * 50 + ['IT'] * 50 + ['Sales'] * 50,
    'Skor Performa' : np.concatenate([hr_scores, it_scores, sales_scores])
})
```

```python
# Analsis Statistik
for dept in ['HR', 'IT', 'Sales']:
    scores = df[df['Departement'] == dept]['Skor Performa']

    mean_value = np.mean(scores)
    std_value = np.std(scores)
    var_value = np.var(scores)

    bound_1sigma = (mean_value - std_value, mean_value + std_value)
    bound_2sigma = (mean_value - 2*std_value, mean_value + 2*std_value)
    bound_3sigma = (mean_value - 3*std_value, mean_value + 3*std_value)

    print(f"Sigma ±1 Std Dev: {bound_1sigma[0]:.2f}%, {bound_1sigma[1]:.2f}%")
    print(f"Sigma ±2 Std Dev: {bound_2sigma[0]:.2f}%, {bound_2sigma[1]:.2f}%")
    print(f"Sigma ±3 Std Dev: {bound_3sigma[0]:.2f}%, {bound_3sigma[1]:.2f}%")
    
    # Aturan Empiris (68%)
    within_1sigma = np.sum((scores >= mean_value - std_value) & 
                           (scores <= mean_value + std_value)) / len(scores) * 100
    print(f"Data dalam ±1 Std Dev: {within_1sigma:.2f}%")

# Visualisasi Histogram
plt.figure(figsize=(10,5))
for dept, color in zip(['HR', 'IT', 'Sales'], ['blue', 'green', 'red']):
    plt.hist(df[df['Departement'] == dept]['Skor Performa'], bins=15, alpha=0.5, label=dept, color=color, edgecolor='black')


# Garis Mean dan batas standar deviasi
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label="Mean")
plt.axvline(bound_1sigma[0], color='green', linestyle='dotted', linewidth=2, label="-1 Std Dev")
plt.axvline(bound_1sigma[1], color='green', linestyle='dotted', linewidth=2, label="+1 Std Dev")
plt.axvline(bound_2sigma[0], color='orange', linestyle='dotted', linewidth=2, label="-2 Std Dev")
plt.axvline(bound_2sigma[1], color='orange', linestyle='dotted', linewidth=2, label="+2 Std Dev")
plt.axvline(bound_3sigma[0], color='purple', linestyle='dotted', linewidth=2, label="-3 Std Dev")
plt.axvline(bound_3sigma[1], color='purple', linestyle='dotted', linewidth=2, label="+3 Std Dev")

plt.legend()
plt.title("Distribusi Skor Performa Karyawan")
plt.xlabel("Skor Performa")
plt.ylabel("Frekuensi")
plt.show()
```
