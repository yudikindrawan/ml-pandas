## Introduction to Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

```python
plt.plot()
plt.show()
```

```python
plt.plot([1, 2, 3, 4])
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y)
plt.show()
```

```python
# first method

fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# second method

fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# third method (recommended)

fig, ax = plt.subplots()
ax.plot(x, y) #add some data
plt.show()
type(fig), type(ax)
```
